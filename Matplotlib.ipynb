```
#importando as bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```
np.random.seed(42)
x= np.arange(10,)
y= np.random.normal(size=10)
```

```
#plotar os valores
plt.plot(x, y)
plt.title('Exemplo')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.show()
```

```
#abordagem orientada a objeto

fig, ax = plt.subplots()
```

```
x = np.arange(10)

fig, ax = plt.subplots()

ax.plot(x, x, label='Linear')
ax.plot(x, x*x, label='Quadrado')
ax.set_title('Abordagem Orientada a Objeto')
ax.set_ylabel('Eixo Y')
ax.set_xlabel('Eixo X')
ax.legend()

plt.show()
```

```
# plotar 2 gráficos lado a lado
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,3))

#grafico 1
ax[0].plot(x,x)
ax[0].set_title('Linear')
ax[0].set_xlabel('Eixo X')
ax[0].set_ylabel('Eixo Y')


#grafico 2
ax[1].plot(x,x*x)
ax[1].set_title('Quadrado')
ax[1].set_xlabel('Eixo X')
ax[1].set_ylabel('Eixo Y')

plt.show()
```

```
#abordagem interface pyplot

x = np.arange(10)

plt.plot(x, x)
plt.plot(x, x**2)
plt.plot(x, x**3)
plt.title('Linear')
plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')

plt.show()
```

```
import pandas as pd
import matplotlib.transforms as mtransforms

url = 'https://fred.stlouisfed.org/graph/fredgraph.csv?id=VIXCLS'
vix = pd.read_csv(url, index_col=0, parse_dates=True, na_values='.',
                  infer_datetime_format=True,
                  squeeze=True).dropna()
ma = vix.rolling('90d').mean()
state = pd.cut(ma, bins=[-np.inf, 14, 18, 24, np.inf],
               labels=range(4))
```

```
x = np.arange(-4, 5)
y1 = x ** 2
y2 = 10 / (x ** 2 + 1)
fig, ax = plt.subplots()
ax.plot(x, y1, 'rx', x, y2, 'b+', linestyle='solid')
ax.fill_between(x, y1, y2, where=y2>y1, interpolate=True,
                color='green', alpha=0.3)
lgnd = ax.legend(['y1', 'y2'], loc='upper center', shadow=True)
lgnd.get_frame().set_facecolor('#ffb19a')
plt.show()
```
