```
#importar bibliotecas
import pandas as pd
import matplotlib.pyplot as plt

#importar dados para dataframe
data = "https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/dengue-dataset.csv"
df = pd.read_csv(data, index_col='data', parse_dates=True)

#ver as 5 primeiras entradas
df.head()
```

```
#plotar grafico orientado a objetos

fig, ax = plt.subplots(figsize=(19,8))



ax.plot('casos-confirmados', data=df)
ax.set_title('Casos de Dengue em Campinas')
ax.set_xlabel('Data')
ax.set_ylabel('Casos Confirmados')
fig.autofmt_xdate()


plt.show()
```

```
#plotar grafico de temperatura média de 2011 a 2013
fig, ax = plt.subplots(figsize=(19,8))

#selecionar intervalo desejado
df_sel = df.loc['2011-01-01':'2013-12-31']


ax.plot('temperatura-media', data=df_sel)
ax.set_title('Casos de Dengue em Campinas')
ax.set_xlabel('Data')
ax.set_ylabel('Casos Confirmados')
fig.autofmt_xdate()
plt.show()
```

```
#plotar multiplos graficos separadamente

#plotar grafico de temperatura media e casos confirmados

fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(19,8))

#temperatura media
ax[0].plot('temperatura-media', data=df)
ax[0].set_title('Temperatura média')
ax[0].grid()

#casos confirmados
ax[1].plot('casos-confirmados', data=df)
ax[1].set_title('Casos Confirmados')
ax[1].grid()

plt.tight_layout()
```
